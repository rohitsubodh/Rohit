```python
train_issue_date = df['issue_date'].astype(str)
train_listing_date = df['listing_date'].astype(str)
test_issue_date = df1['issue_date'].astype(str)
test_listing_date = df1['listing_date'].astype(str)
import datetime
train_data_days = []
test_data_days = []
for (i1,i2) in zip(train_issue_date,train_listing_date):
    s1 = datetime.datetime.strptime(i1,'%Y-%m-%d %H:%M:%S')
    s2 = datetime.datetime.strptime(i2,'%Y-%m-%d %H:%M:%S')
    train_data_days.append((s2.date()-s1.date()).days)
    

for (i3,i4) in zip(test_issue_date,test_listing_date):
    s3 = datetime.datetime.strptime(i3,'%Y-%m-%d %H:%M:%S')
    s4 = datetime.datetime.strptime(i4,'%Y-%m-%d %H:%M:%S')
    test_data_days.append((s4.date()-s3.date()).days)
df['days'] = pd.Series(train_data_days)
df1['days'] = pd.Series(test_data_days)
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
def print_score2(clf,X_train,X_test,y_train,y_test,train=True):
    '''
    print the accuracy score,classification report ,and confusion matrix'''
    if train:
        '''
        training per formance'''
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train,clf.predict(X_train))))
        y_train_pred = cross_val_predict(clf,X_train,y_train,cv=3)
        print("Precision:\t\t{:.4f}".format(precision_score(y_train,
                                                    y_train_pred,
                                                    average='weighted')))
        print("recall:\t\t{:.4f}".format(recall_score(y_train,
                                              y_train_pred,
                                              average='weighted')))
        print("F1:\t\t{:.4f}".format(f1_score(y_train,
                                      y_train_pred,
                                      average='weighted')))
        
    elif train==False:
        '''
        test performance'''
        print("Test Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test,clf.predict(X_test))))
        y_test_pred = cross_val_predict(clf,X_test,y_test,cv=3)
        print("precision score:\t {0:.4f}".format(precision_score(y_test,
                                                          y_test_pred,
                                                          average='weighted')))
        print("Recall score:\t {0:.4f}".format(recall_score(y_test,
                                                    y_test_pred,
                                                    average='weighted')))
        print("F1 score:\t {0:.4f}".format(f1_score(y_test,
                                            y_test_pred,
                                            average='weighted')))
        
```
